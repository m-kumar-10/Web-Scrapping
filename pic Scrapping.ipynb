# Image Scrapping  
first go to image and then copy the source location of that image.

## 1. Setting environment

```python
import requests
img_url = 'https://th.bing.com/th/id/OIP.GEgeMHmS8Xv6k2tBUw40hwHaFj?pid=ImgDet&rs=1'
```

## 2. Get the HTML

```python
html = requests.get(img_url)
html.content  # this will give information in Binary format as we know image are in binary data format.
```

```python
with open('img.jpeg' ,'wb') as ft:
    ft.write(html.content)
```

## 3. And Image has been saved on  Desktop
